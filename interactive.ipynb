```polyglot-notebook
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text.RegularExpressions;

    public class CountryFreelancer
    {
        public string Country { get; set; }
        public int FreelancersCount { get; set; }

        public CountryFreelancer(string country, int count)
        {
            Country = country;
            FreelancersCount = count;
        }

        public override string ToString()
        {
            return string.Concat(Country, ":", FreelancersCount);
        }
    }

    public class ProcessingFreelancers
    {
        public bool ValidateFreelancer(string freelancerData, out string[] parts)
        {
            parts = freelancerData.Split(';')
                .Select(p => p.Trim())
                .ToArray();

            // проверка количества полей
            if(parts.Length != 4)
                return false;

            // проверка фамилии
            var fio = parts[0];
            if(fio.Length < 1 && fio.Length > 40)
                return false;
            for (var i = 0; i < fio.Length; i++)
                if((fio[i])< 65 || (fio[i]) > 90)
                    if((fio[i])< 97 || (fio[i]) > 122)
                        return false;

            // проверка года рождения
            var age = parts[1];
            if(age.Length != 4)
                return false;
            if(!int.TryParse(age, out var ageVar))
                return false;
            if(ageVar < 1960 || ageVar > 2007)
                return false;

            

            // проверка страны
            var country = parts[2];
            if(country.Length < 2 && country.Length > 40)
                return false;
            for (var i = 0; i < country.Length; i++)
                if((int)(country[i])< 65 || (int)(country[i]) > 90)
                    if((int)(country[i])< 97 || (int)(country[i]) > 122)
                        return false;

            

            // проверка ProjectID
            var id = parts[3];
            if(id.Length != 5)
                return false;
            for (var i = 0; i < 2; i++)
                if((int)(id[i])< 65 || (int)(id[i]) > 90)
                    if((int)(id[i])< 97 || (int)(id[i]) > 122)
                        return false;

            // parts[0]="test";
            for (var i = 2; i < 5; i++)
                if((int)(id[i])< 48 || (int)(id[i]) > 57)
                        return false;

            // parts[0]="test";
            return true;
        }

        public IEnumerable<CountryFreelancer> GenerateFreelancersReport(List<string> inputLines)
        {
            var countryCounts = new Dictionary<string, int>();

            foreach (var line in inputLines)
            {
                if (ValidateFreelancer(line, out string[] parts))
                {
                    var country = parts[2];
                    if(!countryCounts.TryGetValue(country, out var count))
                    {
                        countryCounts.Add(country, 1);
                        continue;
                    }
                    countryCounts[country] = count+1;
                }
                Console.WriteLine(parts[0]);
            }

            if (countryCounts.Count == 0)
            {
                countryCounts.Add("none", 0);
            }

            // сортировка
            var sortedCountries = countryCounts.OrderByDescending(key => key.Value);

            foreach (var pair in sortedCountries)
            {
                yield return new CountryFreelancer(pair.Key, pair.Value);
            }
        }
    }
```
